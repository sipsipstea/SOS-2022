```python
import matplotlib.pyplot as plt
from numpy.random import rand
import numpy as np

# make 4 random datasets from 0 to 100
x1 = 100*rand(100)
x2 = 100*rand(100)
y1 = 100*rand(100)
y2 = 100*rand(100)

# plot the data as scatter plot
plt.figure(figsize=(15,12))
plt.axis([0,100,0,100])
splot = plt.scatter(x1,y1, c="blue", s=7)
splot = plt.scatter(x2,y2, c="red", s=7)
plt.show()

# makes a list of tuples, linking the x and y coordinates into one point
set1 = list(zip(x1,y1))
set2 = list(zip(x2,y2))

# sorts the tuples by increasing x coordinate
set1.sort()
set2.sort()

# sorted arrays
x_1 = [m[0] for m in set1]
y_1 = [m[1] for m in set1]
x_2 = [m[0] for m in set2]
y_2 = [m[1] for m in set2]

# plot sorted data connected by lines
plt.figure(figsize=(15,12))
plt.axis([0,100,0,100])
lplot = plt.subplot()
lplot.plot(x_1,y_1, c="blue", label = "Dataset 1")
lplot.plot(x_2,y_2, c="red", label = "Dataset 2")

# fill the area in between curves
xfill = np.sort(np.concatenate([x_1, x_2]))
y1fill = np.interp(xfill, x_1, y_1)
y2fill = np.interp(xfill, x_2, y_2)
lplot.fill_between(xfill, y1fill, y2fill, where=y1fill < y2fill, interpolate=True, color="blue", alpha=0.2)
lplot.fill_between(xfill, y1fill, y2fill, where=y1fill > y2fill, interpolate=True, color="red", alpha=0.2)
plt.show()

# calculate the unsigned area between these 2 lines
i = 0
larea = 0
rarea = 0
while i<199:
    larea = larea + (abs(y1fill[i]-y2fill[i])*(xfill[i+1]-xfill[i]))
    rarea = rarea + (abs(y1fill[i+1]-y2fill[i+1])*(xfill[i+1]-xfill[i]))
    i=i+1

# the unsigned area as an inverse measure of overlapping data
# overlap ranges from 0 (complete overlap) to 100 (least overlap possible)
overlap = (larea+rarea)/200
print(overlap)
```
